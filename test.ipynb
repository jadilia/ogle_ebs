```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = "phot/I/OGLE-GD-ECL-07133.dat"
```

```python
# df = pd.read_table?
```

```python
df = pd.read_table(file,delim_whitespace=True,names=('time','mag','err'))
df
```

```python
p = 7.370563 # days
```

```python
plt.scatter(df['time'], df['mag'], c = df['time'])
plt.xlabel('JD-2450000 [days]')
plt.ylabel('I [mag]')
```

```python
phase = np.mod(df['time'], p) / p


plt.scatter(phase, df['mag'], alpha = 0.5, c = df['time'])
plt.xlabel('phase')
plt.ylabel('I [mag]')
plt.ylim(16.4, 15.6)
plt.xlim(0, 1)
```
